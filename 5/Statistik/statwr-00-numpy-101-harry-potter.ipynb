## 1. Bild einlesen
Digitale Farbbilder sind in Numpy dreidimensionale Arrays mit den Dimensionen *(Höhe,Breite,3)*. Die letzte Dimension entspricht den drei Farbkanälen Rot, Grün und Blau. Die einzelnen Pixelwerte liegen zwischen 0 (dunkel) und 255 (hell). 
Der folgende Code liest ein Bild ein und plottet es. Das Bild zeigt 28 Gesichter aus den Harry-Potter-Filmen.

Bestimmen Sie die Breite und Höhe des Bildes mittels der **shape** des Numpy-Arrays X.

```python
# Imports von Numpy sowie PIL (zum Einlesen von Bildern) und Matplotlib (zum Plotten)
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Lese ein Bild als Numpy-Array ein
def read_jpg(path):
    return np.asarray(Image.open(path))

def plot(X):
    plt.imshow(X, cmap=plt.cm.gray, vmin=0, vmax=255)
    plt.show()

X = read_jpg('/kaggle/input/d/aulges/potterfaces/faces.png')
plot(X)
```
